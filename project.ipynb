# Exploratory Data Analysis(EDA)

```python
# Line plot to show the trend of bitcoin market over time, note: the market price of bitcoin was low throughout the year 2010.
x = df.head(70)['Date']
y = df.head(70)['btc_market_price']
plt.plot(x,y)
plt.title('Bitcoin Market Trend over Time')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.show;
```

```python
# Line plot to show the trend of bitcoin market over time
x = df.tail(70)['Date']
y = df.tail(70)['btc_market_price']
plt.plot(x,y)
plt.title('Bitcoin Market Trend over Time')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=90)
plt.show;
```

```python
# sum of the market price of bitcoins per each year, note: bitcoin market started around 1st of January,2011
df.groupby('Date')['btc_market_price'].sum()
```

```python
# number of daily confirmed Bitcoin transactions over time
x = df.head(70)['Date']
y = df.head(70)['btc_n_transactions']
plt.plot(x,y)
plt.title('Bitcoin Transaction Over Time')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=90)
plt.show;
```

```python
# a correlation between Bitcoin market capitalization and hash rate
plt.scatter(df['btc_market_cap'], df['btc_hash_rate'], c='b', marker='*')
plt.xlabel('btc_market_cap')
plt.ylabel('btc_hash_rate')
plt.title('The Relationship Between Market Capitalization and Hash rate')
plt.grid(True)
plt.show();
```

```python
# the distribution of Bitcoin transaction fees 
plt.hist(df['btc_transaction_fees'], bins=30)
plt.xlabel('range')
plt.ylabel('frequency')
plt.title('Bitcoin Transaction Fees')
plt.show();
```

```python
difficult_in_coin =df.groupby('Date')['btc_difficulty'].sum()
difficult_in_coin
```

```python
# Line plot to show how difficulty it is to find a new block change over year
b = difficult_in_coin.plot()
plt.title('Bitcoin mining difficulty over the year')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=45)
plt.show;
```

```python
#the most active addresses on the bitcoin blockchain
sorted_df = df.sort_values(by='btc_n_unique_addresses', ascending=False)
most_active_addresses= sorted_df.head(10)['btc_n_unique_addresses']
most_active_addresses
```

```python
#the total output volume of bitcoin transaction over time
x = df.head(70)['Date']
y = df.head(70)['btc_output_volume']
plt.plot(x,y)
plt.title('Bitcoin Volume Transaction Over Time')
plt.xlabel('Date')
plt.ylabel('Value')
plt.xticks(rotation=90)
plt.show;
```
